```python
import os
import warnings
warnings.filterwarnings('ignore')

# Creating project folders
directories = ['data/raw', 'data/processed', 'notebooks', 'models', 'outputs']
for directory in directories:
    os.makedirs(directory, exist_ok=True)
```

```python
import gzip
import shutil

try:
    if os.path.exists('data/raw/listings.csv.gz'):
        with gzip.open('data/raw/listings.csv.gz', 'rb') as f_in:
            with open('data/raw/listings.csv', 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)
    else: print("File not present in directory")

    os.remove('data/raw/listings.csv.gz')
except Exception as e:
    print(f'Error: {e}')
```
